```python
import os
import json
import pyarrow.parquet as pq

exp_dir = "Data/Collaborative Expert Two Handed Object Transfer Small"
tool_type = "FENESTRATED_BIPOLAR_FORCEPS:420205"
task_index = 0

info = {
    "dataset_name": "Two Handed Object Transfer",
    "num_episodes": 0,
    "num_steps": 0,
    "observation": {"dtype": "float32", "shape": [14]},
    "action": {"dtype": "float32", "shape": [14]},
    "cameras": ["camera_left.mp4", "camera_right.mp4"],
    "tool_type": tool_type,
    "task_index": task_index,
    "episode_parquet_paths": []
}

# Process demo folders sequentially
demo_folders = sorted([d for d in os.listdir(exp_dir) if d.startswith("Demo")])
for demo in demo_folders:
    demo_path = os.path.join(exp_dir, demo)
    # Find parquet file
    parquet_files = [f for f in os.listdir(demo_path) if f.endswith(".parquet")]
    for pf in sorted(parquet_files):
        parquet_path = os.path.join(demo, pf)  # relative path
        info["episode_parquet_paths"].append(parquet_path)
        # Count steps
        table = pq.read_table(os.path.join(demo_path, pf))
        info["num_steps"] += table.num_rows
        info["num_episodes"] += 1

# Save info.json at exp_dir
with open(os.path.join(exp_dir, "info.json"), "w") as f:
    json.dump(info, f, indent=4)

print(f"Saved info.json with {info['num_episodes']} episodes and {info['num_steps']} steps.")
```
