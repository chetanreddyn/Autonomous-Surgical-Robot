```python
import numpy as np
import pandas as pd

def episode_stats_from_df(df, episode_index):
    out = {"episode_index": int(episode_index), "stats": {}}
    
    if "episode_index" in df.columns:
        df = df[df["episode_index"] == episode_index]
    else:
        print("Episode Index not found in the passed dataframe")
    
    for col in df.columns:
        s = df[col].dropna()
        if s.empty:
            continue
        
        # numeric scalar column
        if pd.api.types.is_numeric_dtype(s):
            vals = s.values.astype(float)
            out["stats"][col] = {
                "min": [float(vals.min())],
                "max": [float(vals.max())],
                "mean": [float(vals.mean())],
                "std": [float(vals.std(ddof=0))],
                "count": [len(vals)]
            }
            continue
        
        # try array-like column (observation.state, action, etc)
        try:
            arrs = [np.asarray(v, dtype=float) for v in s]
            stacked = np.stack(arrs, axis=0)  # shape: (N, ...)
            
            out["stats"][col] = {
                "min": stacked.min(axis=0).tolist(),
                "max": stacked.max(axis=0).tolist(),
                "mean": stacked.mean(axis=0).tolist(),
                "std": stacked.std(axis=0, ddof=0).tolist(),
                "count": [len(stacked)]
            }
        except (ValueError, TypeError):
            # skip non-numeric columns (strings, mixed types, etc)
            continue
    
    return out
```

```python
# python
def compare_stats_dicts(dict1, dict2, tol=1e-6):
    """
    Compare two stats dictionaries. Returns a summary of differences.
    
    Args:
        dict1: First stats dict (can have fewer keys)
        dict2: Second stats dict (reference, can have more keys)
        tol: Tolerance for floating point comparison (default 1e-6)
    
    Returns:
        dict with 'missing_keys', 'extra_keys', 'mismatched_keys', 'matched_keys', and 'all_match' flag
    """
    result = {
        'missing_keys': [],  # keys in dict2 but not in dict1
        'extra_keys': [],    # keys in dict1 but not in dict2
        'mismatched_keys': {},  # keys where values differ beyond tolerance
        'matched_keys': [],  # keys that match perfectly
        'all_match': True
    }
    
    stats1 = dict1.get('stats', dict1)
    stats2 = dict2.get('stats', dict2)
    
    keys1 = set(stats1.keys())
    keys2 = set(stats2.keys())
    
    result['missing_keys'] = sorted(keys2 - keys1)
    result['extra_keys'] = sorted(keys1 - keys2)
    
    if result['missing_keys'] or result['extra_keys']:
        result['all_match'] = False
    
    # Compare common keys
    common_keys = keys1 & keys2
    for key in common_keys:
        val1 = stats1[key]
        val2 = stats2[key]
        
        key_matches = True
        # Compare each stat field (min, max, mean, std, count)
        for stat_name in ['min', 'max', 'mean', 'std', 'count']:
            if stat_name not in val1 or stat_name not in val2:
                continue
            
            arr1 = np.asarray(val1[stat_name], dtype=float).flatten()
            arr2 = np.asarray(val2[stat_name], dtype=float).flatten()
            
            if arr1.shape != arr2.shape:
                if key not in result['mismatched_keys']:
                    result['mismatched_keys'][key] = {}
                result['mismatched_keys'][key][stat_name] = {
                    'reason': 'shape_mismatch',
                    'shape1': arr1.shape,
                    'shape2': arr2.shape
                }
                result['all_match'] = False
                key_matches = False
                continue
            
            # Check if values are within tolerance
            if not np.allclose(arr1, arr2, atol=tol, rtol=0):
                max_diff = np.abs(arr1 - arr2).max()
                if key not in result['mismatched_keys']:
                    result['mismatched_keys'][key] = {}
                result['mismatched_keys'][key][stat_name] = {
                    'reason': 'value_mismatch',
                    'max_diff': float(max_diff),
                    'value1_sample': arr1[:3].tolist() if len(arr1) > 0 else [],
                    'value2_sample': arr2[:3].tolist() if len(arr2) > 0 else []
                }
                result['all_match'] = False
                key_matches = False
        
        if key_matches:
            result['matched_keys'].append(key)
    
    return result

# Test it
comparison = compare_stats_dicts(out['stats'], episode_stats[2]['stats'])
print("Missing keys:", comparison['missing_keys'])
print("Extra keys:", comparison['extra_keys'])
print(f"\nMatched columns ({len(comparison['matched_keys'])}):")
for key in sorted(comparison['matched_keys']):
    print(f"  ✓ {key}")
print("\nAll match:", comparison['all_match'])
if comparison['mismatched_keys']:
    print(f"\nMismatched columns ({len(comparison['mismatched_keys'])}):")
    for key, mismatches in comparison['mismatched_keys'].items():
        print(f"  ✗ {key}:")
        for stat, details in mismatches.items():
            print(f"    {stat}: {details}")
```
